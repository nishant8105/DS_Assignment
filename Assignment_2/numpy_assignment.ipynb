# Numpy Assignment

## 1. Import Libraries
Import the `numpy` library to get started.

```python
import numpy as np
```

## 2. Array Creation
Create a numpy array containing integers from 1 to 20 using `arange`.

```python
arr = np.arange(1, 21)
arr
```

## 3. Basic Statistics
Calculate the sum, mean, median, and standard deviation of the array elements.

```python
print("Sum of array",np.sum(arr))
print("Mean of array",np.mean(arr))
print("Median",np.median(arr))
print("standard deviation of array",np.std(arr))
```

## 4. Conditional Selection
Select elements greater than 10 and find their indices.

```python
print("element greater than 10", arr[arr > 10])
print("Indices of element greater than 10",np.where(arr > 10)[0])
```

## 5. Reshaping Arrays
Create a new array with values 1 to 16 and reshape it into a 4x4 matrix.

```python
arr_4 = np.reshape(np.arange(1, 17), (4, 4))
print("4 * 4 array of range 1 to 16",arr_4)
```

## 6. Transpose
Find the transpose of the 4x4 array.

```python
print("Transpose of array",np.transpose(arr_4))
```

## 7. Summation along Axes
Calculate the sum of elements along rows (axis 0) and columns (axis 1).

```python
print("Sum of rows :", np.sum(arr_4, 0))
print("Sum of Column :", np.sum(arr_4, 1))
```

## 8. Random Array Generation
Generate two 3x3 arrays with random integers between 1 and 20.

```python
arr1 = np.random.randint(1, 21, (3, 3))
arr2 = np.random.randint(1, 21, (3, 3))
print("arr1 : \n", arr1)
print("arr2 : \n", arr2)
```

## 9. Element-wise Arithmetic
Perform addition, subtraction, and multiplication on the two random arrays.

```python
print("Addition\n",arr1 + arr2)
print("Subtraction\n",arr1 - arr2)
print("Multiplication\n",arr1 * arr2)
```

## 10. Matrix Multiplication
Calculate the dot product of the two arrays.

```python
print("Dot product of arr1 and arr2",np.dot(arr1, arr2))
```

## 11. Custom Array Reshaping
Create a 1D array with specific values and reshape it into a 3x4 matrix.

```python
arr_1d = np.array([18, 23, 34 ,56, 43, 78, 87, 45, 67, 16, 31, 94])
arr_2d = np.reshape(arr_1d, (3, 4))
print(arr_2d)
```

## 12. Array Slicing
Extract a sub-array from the 3x4 matrix (first two rows, last two columns).

```python
print(arr_2d[0 : 2, 2 : 4])
```
