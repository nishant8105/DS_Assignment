```python
import numpy as np
```

```python
arr = np.arange(1, 21)
arr
```

```python
print("Sum of array",np.sum(arr))
print("Mean of array",np.mean(arr))
print("Median",np.median(arr))
print("standard deviation of array",np.std(arr))
```

```python
print("element greater than 10", arr[arr > 10])
print("Indices of element greater than 10",np.where(arr > 10)[0])
```

```python
arr_4 = np.reshape(np.arange(1, 17), (4, 4))
print("4 * 4 array of range 1 to 16",arr_4)
```

```python
print("Transpose of array",np.transpose(arr_4))
```

```python
print("Sum of rows :", np.sum(arr_4, 0))
print("Sum of Column :", np.sum(arr_4, 1))
```

```python
arr1 = np.random.randint(1, 21, (3, 3))
arr2 = np.random.randint(1, 21, (3, 3))
print("arr1 : \n", arr1)
print("arr2 : \n", arr2)
```

```python
print("Addition\n",arr1 + arr2)
print("Subtraction\n",arr1 - arr2)
print("Multiplication\n",arr1 * arr2)
```

```python
print("Dot product of arr1 and arr2",np.dot(arr1, arr2))
```
