```python
import pandas as pd
import numpy as np
```

```python
data = {
      'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
      'Age': [24, 27, 22, 32, 29],
      'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
      'Salary': [45000, 54000, 50000, 62000, 47000]
  }

df = pd.DataFrame(data)
print(df.head(5))
```

```python
print(df.describe())
```

```python
hr = df[df['Department'] == 'HR']
print("Average Salary of HR : ",np.mean(hr['Salary']))
```
