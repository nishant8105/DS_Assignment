# Matplotlib Assignment

## 1. Import Libraries
Import `matplotlib.pyplot` and `numpy`.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Line Plot
Define x and y data points and create a basic line plot.

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
```

```python
plt.plot(x, y, color='r', marker='o')
plt.legend('X, Y')
plt.show()
```

## 3. Enhanced Line Plot
Create a line plot with grid, title, and axis labels.

```python
plt.plot(x, y, color='r', marker='o')
plt.legend('X, Y')
plt.grid()
plt.title('X vs Y')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

## 4. Bar Chart
Create a bar chart showing student marks.

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
```

```python
plt.bar(students, marks , color = ['red', 'blue', 'purple', 'green'])
plt.title("Student marks")
plt.show()
```

## 5. Pie Chart
Create a pie chart showing revenue distribution by region, with the highest revenue slice exploded.

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
max_revenue = revenue.index(max(revenue))
explode = [0]*len(revenue)
explode[max_revenue] = 0.1
```

```python
plt.pie(revenue,labels= regions, colors= ['red', 'blue', 'purple', 'green'], autopct='%1.1f%%', explode=explode)
plt.show()
```

## 6. Histogram
Generate 1000 random integers between 1 and 100, and create a histogram to show the frequency distribution.

```python
rand_int = np.random.randint(1, 100, 1000)
plt.hist(rand_int, bins=15)
plt.xlabel('value')
plt.ylabel('Frequency')
plt.title("Histogram")
plt.show()
```
